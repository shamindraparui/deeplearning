```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()
```

```python
print("Shape of train-images {0} and shape of train-labels {1}".format(train_images.shape, train_labels.shape))
```

```python
print("Shape of test-images {0} and shape of test-labels {1}".format(test_images.shape, test_labels.shape))
```

```python
plt.imshow(train_images[1002])
plt.show()
print("Corresponding label is {0}".format(train_labels[1002]))
```

```python
print("Data type of the train-images reference is {0}".format(type(train_images)))
```

```python
train_images = train_images.reshape((60000, 28*28))
train_images = train_images.astype("float32") / 255

test_images = test_images.reshape((10000, 28*28))
test_images = test_images.astype("float32") / 255
```

```python
print("Shape of train-images {0} and shape of train-labels {1}".format(train_images.shape, train_labels.shape))
```

```python
train_labels = tf.keras.utils.to_categorical(train_labels)
test_labels = tf.keras.utils.to_categorical(test_labels)
```

```python
test_labels.shape
```

```python
network = tf.keras.models.Sequential()
network.add(tf.keras.layers.Dense(512, activation='relu', input_shape=(28*28,)))
network.add(tf.keras.layers.Dense(10, activation='softmax'))
```

```python
network.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
result = network.fit(train_images, train_labels, epochs=5, batch_size=128)
```

```python
prediction = network.predict(test_images[0:1])
print("Prediction is ",np.argmax(prediction), " and original is ", test_labels[0])
```
