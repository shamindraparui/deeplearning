```python
from tensorflow import keras
import keras_cv
import matplotlib.pyplot as plt
import cv2
```

```python
keras.mixed_precision.set_global_policy("float32")
network = keras_cv.models.StableDiffusion(jit_compile=True, img_width = 512, img_height = 512)
```

```python
images = network.text_to_image("An old Guiter hanging on the wall with climbing plants, \
 apocalypse environment, fantasy art", 4)
```

```python
for img in images:
  plt.imshow(img)
  plt.show()
```

```python
cv2.imwrite('Generated.png', cv2.cvtColor(images[], cv2.COLOR_BGR2RGB))
```
